```python
import numpy as np
from matplotlib import pyplot as plt

plt.figure(figsize=[12, 8], dpi=120)
plt.rcParams.update(plt.rcParamsDefault)
# plt.rcParams['font.sans-serif'] = ['DengXian'] #用来正常显示中文标签
# print(plt.rcParams.get('font.sans-serif'))
# plt.rcParams.update({"figure.figsize": (12, 8), "figure.dpi": 200})
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig1.set_facecolor('white')
fig2.set_facecolor('white')

data = [
[0.4,50,259,9.239,192,10.22],
[0.8,50,251,16.33,168,16.36],
[1.6,50,257,30.75,183,28.91],
[4,50,254,72.06,174,68.64],
[4,20,252,74.83,183,76.00],
[4,50,262,114.9,186,117.8],
[4,100,257,160.7,172,166.9],
[4,125,256,177.9,170,185.8],
[4,150,256,192.5,168,202.7],
[4,175,259,191.6,161,206.9],
[4,200,263,198.5,159,216.8]
]
dataT = np.array(data).T
dotPolyX = np.arange(0.4, 4.0, 1e-3)

ax1.plot(dataT[0][0: 4], dataT[3][0: 4], 'r.', label="Ipa")
ax1.plot(dataT[0][0: 4], dataT[5][0: 4], 'k.', label="Ipc")
dotPolyF = np.poly1d(np.polyfit(dataT[0][0: 4], dataT[3][0: 4], 1))
# ax1.plot(dotPolyX, dotPolyF(dotPolyX), 'r--', linewidth=.9, label="R = {:.4f}".format(dotPolyF.r[0]))
ax1.plot(dotPolyX, dotPolyF(dotPolyX), 'r--', linewidth=.9, label=dotPolyF)
dotPolyF = np.poly1d(np.polyfit(dataT[0][0: 4], dataT[5][0: 4], 1))
ax1.plot(dotPolyX, dotPolyF(dotPolyX), 'k--', linewidth=.9, label="R = {:.4f}".format(dotPolyF.r[0]))
ax1.xaxis.label.set_text(r"Concentration (mmol/L)")
ax1.yaxis.label.set_text(r"Current (μA)")
ax1.set_title(u"Ipa、Ipc 与相应浓度 c 的关系曲线", family='SimHei')
ax1.legend()

dotPolyX = np.arange(np.sqrt(20), np.sqrt(200), 1e-2)
ax2.plot(np.sqrt(dataT[1][4: -1]), dataT[3][4: -1], 'r.', label="Ipa")
ax2.plot(np.sqrt(dataT[1][4: -1]), dataT[5][4: -1], 'k.', label="Ipc")
dotPolyF = np.poly1d(np.polyfit(np.sqrt(dataT[1][4: -1]), dataT[3][4: -1], 1))
# ax2.plot(dotPolyX, dotPolyF(dotPolyX), 'r--', linewidth=.9, label="R = {:.4f}".format(dotPolyF.r[0]))
ax2.plot(dotPolyX, dotPolyF(dotPolyX), 'r--', linewidth=.9, label=dotPolyF)
dotPolyF = np.poly1d(np.polyfit(np.sqrt(dataT[1][4: -1]), dataT[5][4: -1], 1))
ax2.plot(dotPolyX, dotPolyF(dotPolyX), 'k--', linewidth=.9, label="R = {:.4f}".format(dotPolyF.r[0]))
ax2.xaxis.label.set_text(r"Square root of Velocity (mV/s)^(1/2)")
ax2.yaxis.label.set_text(r"Current (μA)")
ax2.set_title(u"Ipa、Ipc 与相应速度 v 的关系曲线", family='SimHei')
ax2.legend()
```
